```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 创建一个空白图像
image = np.zeros((500, 500, 3), dtype=np.uint8)

# 定义矩形的长和宽
rect_width = 200
rect_height = 100

# 矩形的左上角顶点坐标
rect_x, rect_y = 150, 200

# 绘制矩形
cv2.rectangle(image, (rect_x, rect_y), (rect_x + rect_width, rect_y + rect_height), (255, 0, 0), 2)

# 矩形中心点坐标
center_x, center_y = rect_x + rect_width // 2, rect_y + rect_height // 2

# 矩形的长对角线作为椭圆的长轴
ellipse_major_axis = int(np.sqrt(rect_width**2 + rect_height**2))
ellipse_minor_axis = rect_height  # 短轴设为矩形的高度

# 绘制椭圆
cv2.ellipse(image, (center_x, center_y), (ellipse_major_axis // 2, ellipse_minor_axis // 2), 0, 0, 360, (0, 255, 0), 2)

# 使用Matplotlib在Jupyter中显示图像
plt.figure(figsize=(8, 8))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Rectangle with Inscribed Ellipse')
plt.axis('off')  # 关闭坐标轴
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 创建一个空白图像
image = np.zeros((500, 500, 3), dtype=np.uint8)

# 定义矩形的长和宽
rect_width = 200
rect_height = 100

# 矩形的左上角顶点坐标
rect_x, rect_y = 150, 200

# 绘制矩形
cv2.rectangle(image, (rect_x, rect_y), (rect_x + rect_width, rect_y + rect_height), (255, 0, 0), 2)

# 矩形中心点坐标
center_x, center_y = rect_x + rect_width // 2, rect_y + rect_height // 2

# 矩形的对角线长度（椭圆的长轴）
ellipse_major_axis = int(np.sqrt(rect_width**2 + rect_height**2) * 0.9)  # 缩短一些以确保不超出矩形
ellipse_minor_axis = int(min(rect_width, rect_height) * 0.8)  # 确保椭圆不超出矩形

# 计算椭圆的旋转角度（矩形对角线的角度）
angle = np.degrees(np.arctan2(rect_height, rect_width))

# 绘制椭圆
cv2.ellipse(image, (center_x, center_y), (ellipse_major_axis // 2, ellipse_minor_axis // 2), angle, 0, 360, (0, 255, 0), 2)

# 使用Matplotlib在Jupyter中显示图像
plt.figure(figsize=(8, 8))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Rectangle with Inscribed Ellipse')
plt.axis('off')  # 关闭坐标轴
plt.show()
```
